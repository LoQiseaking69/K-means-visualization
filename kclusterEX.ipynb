# K-Means Clustering Example
This notebook demonstrates the implementation and visualization of the K-Means clustering algorithm, a widely used method in unsupervised machine learning for grouping data into clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

## Data Generation
First, we will generate some synthetic data with distinct clusters using `make_blobs` from `sklearn.datasets`.

```python
# Generate synthetic data
n_samples = 300
n_features = 2
n_clusters = 4
random_state = 42

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)
```

## Applying K-Means
We will apply the K-Means clustering algorithm to our synthetic data, trying to group it into clusters.

```python
# Apply K-Means with explicit n_init
n_init_value = 10  # or any other appropriate value
kmeans = KMeans(n_clusters=n_clusters, n_init=n_init_value, random_state=random_state)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

## Visualization of Clusters
Let's visualize the resulting clusters along with their centroids.

```python
# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=100)  # Cluster centroids
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Analysis
The plot above shows the clusters identified by K-Means, with each cluster's centroid marked with a red 'x'. This visualization helps in understanding how K-Means algorithm partitions the data into distinct groups based on their features.

# Conclusion
In this notebook, we used the K-Means clustering algorithm to group synthetic data into clusters and visualized the results. This process demonstrates the effectiveness of K-Means in identifying groupings in datasets.
